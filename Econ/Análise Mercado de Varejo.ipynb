O código abaixo, por sua vez, analisa o Beta de cada ticker, comparando-os com o S&P 500, um indicador do mercado mais amplo. O Beta mede a volatilidade de um ativo em relação ao mercado; um Beta maior que 1 indica que o ativo é mais volátil que o mercado, enquanto um Beta menor que 1 sugere menor volatilidade. No gráfico resultante, observa-se a variação do Beta entre os ativos de e-commerce e varejo físico. Esta análise é vital para entender como os diferentes ativos reagem a movimentos de mercado, auxiliando investidores a compreender o risco relativo de cada ativo em relação às tendências gerais do mercado.

```python
# Cálculo do Beta para cada ticker
market_data = yf.download('^GSPC', start="2010-01-01", end="2021-12-31")  # S&P 500 como mercado
plt.figure(figsize=(15, 8))
for ticker in e_commerce_tickers + retail_tickers:
    data = yf.download(ticker, start="2010-01-01", end="2021-12-31")
    data['Return'] = data['Close'].pct_change()
    market_data['Market_Return'] = market_data['Close'].pct_change()
    covariance = data['Return'].cov(market_data['Market_Return'])
    variance = market_data['Market_Return'].var()
    beta = covariance / variance
    plt.bar(ticker, beta)
plt.title('Beta do Mercado para Cada Ticker')
plt.xlabel('Ticker')
plt.ylabel('Beta')
plt.show()
```

O Índice de Força Relativa (RSI) é um indicador técnico utilizado para identificar a força e a reversão potencial das tendências de preço, operando na faixa de 0 a 100. Valores acima de 70 indicam sobrecompra (potencialmente sobrevalorizado), enquanto valores abaixo de 30 sugerem sobrevenda (potencialmente subvalorizado). Através do cálculo do RSI para cada ticker, o gráfico mostra a variação deste índice ao longo do tempo para os ativos de e-commerce e varejo físico. A partir da análise do gráfico, pode-se identificar períodos onde os ativos estavam potencialmente sobrecomprados ou sobrevendidos, oferecendo insights sobre os momentos ideais para compra ou venda.

```python
# Cálculo do RSI para cada ticker
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    up = delta.clip(lower=0)
    down = -1 * delta.clip(upper=0)
    ema_up = up.ewm(com=window - 1, adjust=False).mean()
    ema_down = down.ewm(com=window - 1, adjust=False).mean()
    rs = ema_up / ema_down
    rsi = 100 - (100 / (1 + rs))
    return rsi

plt.figure(figsize=(15, 8))
for ticker in e_commerce_tickers + retail_tickers:
    data = yf.download(ticker, start="2010-01-01", end="2021-12-31")
    data['RSI'] = calculate_rsi(data)
    plt.plot(data.index, data['RSI'], label=ticker)
plt.title('Índice de Força Relativa (RSI)')
plt.xlabel('Data')
plt.ylabel('RSI')
plt.legend()
plt.show()
```

# 4. Análise Estatística e sobre crescimento de cada setor

Este código calcula a taxa de crescimento anualizada para cada ativo, fornecendo uma métrica de comparação do crescimento a longo prazo entre os setores.

```python
# Cálculo do crescimento anualizado
def annualized_growth(data, years=12):
    return data.iloc[-1] / data.iloc[0] ** (1/years) - 1

growth_rates = {}
for ticker in e_commerce_tickers + retail_tickers:
    data = yf.download(ticker, start="2010-01-01", end="2021-12-31")['Close']
    growth_rates[ticker] = annualized_growth(data)

# Exibir taxas de crescimento
for ticker, growth in growth_rates.items():
    print(f'{ticker}: {growth:.2%}')
```

A análise de autocorrelação revela dependências temporais nos preços das ações, crucial para entender o comportamento temporal dos ativos.

```python
from statsmodels.graphics.tsaplots import plot_acf

# Análise de autocorrelação para verificar dependências temporais
for ticker in e_commerce_tickers + retail_tickers:
    data = yf.download(ticker, start="2010-01-01", end="2021-12-31")['Close']
    plot_acf(data, lags=50, title=f'Autocorrelação para {ticker}')
    plt.show()
```

```python
# Cálculo do rendimento médio para cada setor
rendimentos_ecommerce = pd.DataFrame()
rendimentos_retail = pd.DataFrame()

for ticker in e_commerce_tickers:
    rendimentos_ecommerce[ticker] = yf.download(ticker, start="2010-01-01", end="2021-12-31")['Close'].pct_change()

for ticker in retail_tickers:
    rendimentos_retail[ticker] = yf.download(ticker, start="2010-01-01", end="2021-12-31")['Close'].pct_change()

print("Rendimento Médio E-Commerce:", rendimentos_ecommerce.mean().mean())
print("Rendimento Médio Varejo Físico:", rendimentos_retail.mean().mean())
```


O "T-Statistic" resultante do Teste T de Student é uma medida que indica a diferença entre as médias dos dois grupos em unidades de erro padrão. Um T-Statistic alto sugere que as médias dos grupos são significativamente diferentes. O "P-Value" associado indica a probabilidade de observar uma diferença tão grande ou maior entre as médias, se na realidade elas fossem iguais. Um P-Value baixo (geralmente menor que 0,05) sugere que as diferenças observadas são estatisticamente significativas. Essas medidas ajudam a determinar se há uma diferença significativa nos rendimentos entre os setores de e-commerce e varejo físico.

```python
from scipy.stats import ttest_ind

# Teste T para comparar as médias dos dois setores
t_stat, p_val = ttest_ind(rendimentos_ecommerce.dropna().mean(), rendimentos_retail.dropna().mean())
print("T-Statistic:", t_stat)
print("P-Value:", p_val)
```
